# Count of colleges by region

```python
facu_por_estado = data['state'].value_counts()
```

```python
facu_por_regiao = data['region'].value_counts().reset_index()
facu_por_regiao.columns = ['Region','Number of Colleges']
```

# Plot Bar Chart

```python
import seaborn as sns
plt.figure(figsize = (15,10))
barras = sns.barplot(x = facu_por_estado.values, y = facu_por_estado.index, palette = 'viridis')

for bar in barras.patches:
    plt.text(
        bar.get_width(),                             # Gets the width of the bar to position the text at the end of it
        bar.get_y() + bar.get_height() /2,             # Gets the Y position of the bar to position the text in the center vertically
        f'{int(bar.get_width())}',                   # The bar width value is the value to be displayed
        va = 'center'                                # Align text vertically in the center
    )   
        
plt.title("Number of Colleges by State")
plt.xlabel("Number of Faculties")
plt.ylabel("state")
plt.show()
```

# Count of colleges by regionn

```python
# Count of colleges by region

facu_por_regiao = data['region'].value_counts().reset_index()
facu_por_regiao.columns = ['Region', 'Number of Colleges']

# Transforming to a DataFrame with index    

heatmap_data_simple = facu_por_regiao.set_index('Region')

# Creating a simple heatmap

plt.figure(figure=(8,6))
sns.heatmap(heatmap_data_simple.T, annot=True, fmt='d', cmap="viridis")
plt.title('Simple Heat Map of College Count by Region')
plt.xlabel('region')
plt.ylabel('Number of Faculties')
plt.show()  
```


# Cross count of colleges by region and division

```python
# Cross count of colleges by region and division

heatmap_data_cross = pd.crosstab(data['region'], data['division'])

# Creating a cross heat map

plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data_cross, annot=True, fmt='d', cmap='viridis')
plt.title('Cross-Heat Map of College Count by Region and Division')
plt.xlabel('NCAA Division')
plt.ylabel('Region')
plt.show()
```
